```python
# Se crea una nueva base de datos donde el indice es la primera columna de la primera base
df1 = pd.read_excel('Detalle precios y productos fabricados 2022.xlsx', index_col=2)
df1
```

```python
# Se muestran todos los margenes totales de las ventas de Aaron Nuño Cordova
filtro5 = df1.loc[["AARON NUÑO CORDOVA"],["MARGEN_TOTAL_CALCULADO"]]
filtro5
```

```python
# Se muestran todas las ventas que en cantidad sean mayores a 10
filtro6 = df[df["CANT"] > 10]
filtro6
```

```python
# Se muestran las ventas mayores a 10 en cantidad y que hayan sido a publico en general
filtro7 = df[(df["CANT"] > 10) & (df["NOMBRE_CLIENTE"] == "PÚBLICO EN GENERAL")]
filtro7
```

```python
# Se muestran los precios unitarios menores o igual a 400 o donde el cliente haya sido Luis Enrique Patiño Perez
filtro8 = df[(df["PRECIO_UNITARIO"] >= 400) | (df["NOMBRE_CLIENTE"] == "LUIS ENRIQUE PATIÑO PEREZ")]
filtro8
```

```python
# Se mostraran todas las ventas que su cantidad no sean mayores a 6 y aquellas donde no se haya vendido al publico en general
filtro9 = df[~(df["CANT"] > 6) & ~(df["NOMBRE_CLIENTE"] == "PÚBLICO EN GENERAL")]
filtro9
```
