```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_excel('Datos de Facturación.xlsx')
df
```

```python
df.head()
```

```python
# Reemplazo de datos a string tipo numerico
df.CVE_CLPV = df.CVE_CLPV.replace({'O':'0'}, regex=False)
df.CVE_CLPV = df.CVE_CLPV.replace({'6781|':'6781'}, regex=False)
df.CVE_CLPV = df.CVE_CLPV.replace({'3612C':'3612'}, regex=False)
df.CVE_CLPV = df.CVE_CLPV.replace({'      ':'0'}, regex=False)
df.CVE_CLPV = df.CVE_CLPV.replace({'2385A':'2385'}, regex=False)
df.CVE_CLPV = df.CVE_CLPV.replace({'2510A':'2510'}, regex=False)
df.CVE_CLPV = df.CVE_CLPV.replace({'8440Q':'8440'}, regex=False)
df.CVE_CLPV = df.CVE_CLPV.replace({'M':'0'}, regex=False)
```

```python
df["CVE_CLPV"] = df['CVE_CLPV'].astype(float)
```

```python
df["FECHA_ENT"] = df['FECHA_ENT'].astype(str)
```

```python
df.info()
```

```python
# CVE_CLPV (Valores: 1000 a 2000)
filtro1_1 = df[(df["CVE_CLPV"] > 1000) & (df["CVE_CLPV"] <= 2000)]
filtro1_1
filtro1_1.to_csv("Filtro1_1.csv")
from google.colab import files
files.download("Filtro1_1.csv")
```

```python
# CVE_VEND (Todas las claves excepto “5” y “4”)
filtro2_1 = df[~(df['CVE_VEND']==5) & ~(df['CVE_VEND']==4)]
filtro2_1
filtro2_1.to_csv("Filtro2_1.csv")
from google.colab import files
files.download("Filtro2_1.csv")
```

```python
# FECHA_ENT (Fechas de “28/02/2022”)
filtro3_1 = df[df["FECHA_ENT"] == "2022-04-13 00:00:00"]
filtro3_1
filtro3_1.to_csv("Filtro3_1.csv")
from google.colab import files
files.download("Filtro3_1.csv")
```

```python
# CAN_TOT (Cantidades menores a 5951.7)  o  STATUS (“E”)
filtro4_1 = df[(df["CAN_TOT"] < 5951.7) | (df["STATUS"] == "E")]
filtro4_1
filtro4_1.to_csv("Filtro4_1.csv")
from google.colab import files
files.download("Filtro4_1.csv")
```

```python
# Solo las columnas: CVE_DOC, FECHA_ENT, FECHA_VEN y CAN_TOT
filtro5_1 = df.iloc[:,[0,6,7,9]]
filtro5_1
filtro5_1.to_csv("Filtro5_1.csv")
from google.colab import files
files.download("Filtro5_1.csv")
```

```python
# Solo las filas de 7001-7099
filtro6_1 = df.iloc[7001:7100, :]
filtro6_1
filtro6_1.to_csv("Filtro6_1.csv")
from google.colab import files
files.download("Filtro6_1.csv")
```
