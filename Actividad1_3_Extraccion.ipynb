```python
# Eliminar la fila “0” de  “Unnamed” y Eliminar la columna “0” y la columna  “Referencia”
df=pd.read_excel('Inmuebles.xlsx', skiprows=1)
df
```

```python
df = df.drop(['Referencia'], axis=1)
df
```

```python
df
```

```python
# Todos las registros de “Fecha Alta : igual a 2007-04-29 00:00:00 y a 2004-01-02 00:00:00
filtro1_2 = df[(df["Fecha_Alta"] == '2007-04-29') | (df["Fecha_Alta"] == "2004-01-02")]
filtro1_2
filtro1_2.to_csv("Filtro1_2.csv")
from google.colab import files
files.download("Filtro1_2.csv")
```

```python
# Todos los registros de “ Tipo:  parking, local, industrial y piso”
filtro2_2 = df[(df["Tipo"] == "Parking") | (df["Tipo"] == "Local") | (df["Tipo"] == "Industrial") | (df["Tipo"] == "Piso")]
filtro2_2
filtro2_2.to_csv("Filtro2_2.csv")
from google.colab import files
files.download("Filtro2_2.csv")
```

```python
# Todos los registros con “Provincia” igual a “Girona” y con “Superficie” mayor a “100”
filtro3_2 = df[(df["Provincia"] == "Girona") & (df["Superficie"] > 100)]
filtro3_2
filtro3_2.to_csv("Filtro3_2.csv")
from google.colab import files
files.download("Filtro3_2.csv")
```

```python
# Todos los registros con “Precio Venta: mayor a 1000,000” o “vendedor” igual a  “Carmen o Joaquin”
filtro4_2 = df[(df["Precio Venta"] > 1000000) | (df["Vendedor"] == "Carmen") | (df["Vendedor"] == "Joaquín")]
filtro4_2
filtro4_2.to_csv("Filtro4_2.csv")
from google.colab import files
files.download("Filtro4_2.csv")
```

```python
# Todos los registros que no sean de “Provincia : Barcelona“
filtro5_2 = df[~(df["Provincia"] == "Barcelona")]
filtro5_2
filtro5_2.to_csv("Filtro5_2.csv")
from google.colab import files
files.download("Filtro5_2.csv")
```

```python
# Todos los registros que no sean de “Operación: alquiler “ ni “Tipo: oficina“
filtro6_2 = df[~(df["Operación"] == "alquiler") & ~(df["Tipo"] == "oficina")]
filtro6_2
filtro6_2.to_csv("Filtro6_2.csv")
from google.colab import files
files.download("Filtro6_2.csv")
```
